```python
columns = ['cement', 'blast_furnace_slag', 'water', 'superplasticizer','coarse_aggregate', 'fine_aggregate ', 'age','concrete_compressive_strength']
```

```python
for col in columns:
    plt.figure(figsize=(6,4))
    sns.histplot(data[col],kde = True, bins = 30)
    skewness = data[col].skew()
    plt.title(f"Distribution of {col}\n Skew : {skewness:.2f}")
    plt.show()
    if abs(skewness) > 0.80:
        plt.figure(figsize=(6,4))
        sns.histplot(np.log1p(data[col]), kde=True, bins=30)
        plt.title(f"Log-Transformed Distribution of {col}")
        plt.show()
```

```python
corr = data.corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```
